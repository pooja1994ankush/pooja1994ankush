# 2) Cleaning Data :- 

       Dataset contain missing data we have to fill that data or we have to drop that data

### Rating has 1474 missing values by scatter plot we can see outliers, if there is no outliers present in the  graph          then we can fill the missing values by mean

```python
plt.figure(figsize=(16,30))
sns.scatterplot(x=df['Rating'],y=df['Genres'])
plt.show()
```

```python
'''
Filling data
'''
df['Rating'].mean()
```

```python
df['Rating']=df['Rating'].fillna(df['Rating'].mean())
df
```

###  here we are not require last updated, Current ver,Android ver,Generes columns so we will drop it

```python
df.drop(columns={'Genres','Last Updated','Current Ver','Android Ver','Type'},inplace=True)
```

# 3)  Exploratory Data Analysis

    corr() : It returns correlation. 
   

```python
df.corr()
```

```python
df.describe()
```

```python
# heat map

sns.heatmap(df.corr(),annot=True)
```

```python
# App for specific user vs No. of Apps

plt.figure(figsize=(10,5))
sns.countplot(x=df['Content Rating'],order=df['Content Rating'].value_counts().index.sort_values())

plt.xticks(rotation='vertical')
plt.xlabel('App Users',fontsize=12)
plt.ylabel('No. of Apps',fontsize=12)
plt.title('App Users vs No. of Apps',fontsize=16)
```

###   By above graph I concluded that playstore has more than 8000 apps which can use everyone 

```python
# Category vs No. of Apps

plt.figure(figsize=(16,7))
sns.countplot(x=df['Category'],order=df['Category'].value_counts().index.sort_values(),palette='dark')
plt.xlabel('App Category',fontsize=12)
plt.ylabel('No.of Apps',fontsize=12)
plt.xticks(rotation='vertical')
```

### By above graph we can say that playstore has more than 1900 apps which are family related

```python
# Category vs Rating 
plt.figure(figsize=(16,7))
sns.barplot(x=df['Category'].sort_values(),y=df['Rating'])
plt.xticks(rotation='vertical')
plt.xlabel('Category',fontsize=12)
plt.ylabel('Ratings',fontsize=12)
```
